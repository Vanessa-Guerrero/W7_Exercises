```python
import numpy as np
import matplotlib.pyplot as plt 
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5] 
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, marker='o')
plt.show()
```

### Styling Plots

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, ls='--', c='r', marker='o')
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, ls='--', c='r', marker='s', ms=10, mfc='g')
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, 'h-.c', ms=7)
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, 's--r', ms=10, mfc='k', mec='g')
plt.show()
```

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D'] 
sales = [23, 17, 35, 29] 
```

```python
plt.bar(products, sales)
plt.show() 
```

```python
plt.title("Sales by Product")
plt.xlabel('Products')
plt.ylabel('Sales')

plt.bar(products,  
        sales,  
        color='Teal',  
        width=.5) 
plt.show() 
```

```python
x = np.random.rand(50) 
y = np.random.rand(50) 
plt.scatter(x, y)
plt.show()
```

```python
colors = np.random.rand(50)
sizes= np.random.rand(50) * 100
plt.scatter(x, y, c=colors, s=sizes, cmap='viridis') 
plt.colorbar()
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

x = np.linspace(0, 10, 100) 
y = np.sin(x)

axes[0].plot(x, y, color='blue', linestyle='-') 
plt.show()  
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

x = np.linspace(0, 10, 100) 
y = np.sin(x)

axes[0].plot(x, y, color='m')


nums = np.random.randn(1000) 

axes[1].hist(nums, bins=10, alpha=.5, color='g')

plt.show()  
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

x = np.linspace(0, 10, 100) 
y = np.sin(x)

axes[0].plot(x, y, color='m')
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')


nums = np.random.randn(1000) 

axes[1].hist(nums, bins=10, alpha=.5, color='g')
axes[1].grid(axis='y', alpha=.3)
axes[1].set_title('Histogram of Random Numbers')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

plt.show()  
```
