```python
import numpy as np
import matplotlib.pyplot as plt 
```

### Creating Basic Plots

```python
x = [1, 2, 3, 4, 5] 
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, marker='o')
plt.show()
```

### Styling Plots

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, ls='--', c='r', marker='o')
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, ls='--', c='r', marker='s', ms=10, mfc='g')
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, 'h-.c', ms=7)
plt.show()
```

```python
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y, 's--r', ms=10, mfc='k', mec='g')
plt.show()
```
