```python
import numpy as np
import pandas as pd
```

### Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))  
# won't run as a set; produces an error since Series requires an ordered data structure and sets are unordered
ser4 = pd.Series(ad)

print(f"{ser1}\n \n{ser2}\n \n{ser3}\n \n{ser4}")
```

```python
ser5 = pd.Series([17, 18, 19, 20])
print(ser5)
```

### DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 1, ser2)
df1.insert(2, 2, ser3)
df1
```

```python
df1.insert(3, 3, ser4)
df1
```

```python
# adding ser4 into column 3 produced an error because it's a dictionary
# the dictionary keys become the row labels and since the df has row labels as indexes(int)
# the spelled out keys(labels) from the dictionary won't match into the dataframe
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```
